```python
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic=sns.load_dataset("titanic")
planets=sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size")
```

```python
sns.boxplot(x="sex", y="total_bill", data=tips)
```

```python
sns.histplot(tips['tip'], kde=True, bins =15)
```

```python
sns.barplot(x="sex", y="tip", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True)
```

```python
sns.jointplot(x="tip", y="total_bill",data =tips, kind="hex", cmap="Spectral")
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue="pclass")
```
